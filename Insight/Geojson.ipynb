```python
import geopandas as gpd
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
data = gpd.read_file('bairros_fortaleza.json')
```

```python
data.head()
```

```python
ax = data.plot(color='blue')
```

```python
def bairro(str):
    bairro = data[data['BAIRROS'] == str]
    
    return bairro
```

```python
teste = bairro('Aldeota')
ax = teste.plot()
```
