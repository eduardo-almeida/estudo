```python
import numpy
from sklearn.preprocessing import normalize
```

```python
def predict(row, coefficients):
  yhat = coefficients[0]
  for i in range(len(row)-1):
    yhat += coefficients[i + 1] * row[i]
  return yhat
```

```python
def coefficients_sgd(train, l_rate, n_epoch):
    coef = [0.0 for i in range(len(train[0]))]
    print ('Coeficiente Inicial={0}' % (coef))
    for epoch in range(n_epoch):
        sum_error = 0
        for row in train:
            yhat = predict(row, coef)
            error = yhat - row[-1]
            sum_error += error**2
            coef[0] = coef[0] - l_rate * error
            for i in range(len(row)-1):
                coef[i + 1] = coef[i + 1] - l_rate * error * row[i] 
        print(('epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error)))
    return coef
```

```python
def load_csv(file):
    x = []
    y = []
    for line in file:
        x.append((line.split(';')[0:11]))
        y.append((line.split(';')[11]))
    return x, y
```

```python
x, y = load_csv(open('winequality-white.csv'))
```

```python
del x[0]
del y[0]
```

```python
x
```

```python
for i in range(len(x)):
    for j in range(len(x[i])):
        x[i][j] = float(x[i][j])
```

```python
for i in range(len(y)):
    y[i] = float(y[i])
```

```python
dataset = x

for i in range(len(dataset)):
    dataset[i].append(y[i])
```

```python
normed_matrix = normalize(numpy.array(dataset), axis=1, norm='l1')
```

```python
l_rate = 0.001
n_epoch = 10
coef = coefficients_sgd(normed_matrix, l_rate, n_epoch)
print(coef)
```
