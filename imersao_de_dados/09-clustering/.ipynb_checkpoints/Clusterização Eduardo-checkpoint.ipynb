```python
import numpy as np 
# data processing
import pandas as pd 
# data visualization
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv('dataset.csv', header=None)
dataset = np.array(dataset)

plt.scatter(dataset[:,0], dataset[:,1], s=10)
plt.show()
```

```python
def calculate_initial_centers(dataset, k):
    lista =[]
    for i in range(k):
        lista.append(dataset[k])
    return lista
```

```python
centroids = calculate_initial_centers(dataset,3)
```
