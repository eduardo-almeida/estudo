```python
# import das bibliotecas importantes
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
iris_data, rotules = [iris.get('data'), iris.get('target')]
print(iris_data)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(iris_data)
```

```python
predictions = kmeans.predict(iris_data)

reduced_data = PCA(n_components=2).fit_transform(iris_data)

reduced_data_df = pd.DataFrame(reduced_data)
reduced_data_df[2] = predictions

# k-means clusters
fig, ax = plt.subplots()
ax.scatter(reduced_data_df[0], reduced_data_df[1], c=reduced_data_df[2])
plt.show()

# Original
fig, ax = plt.subplots()
ax.scatter(reduced_data_df[0], reduced_data_df[1], c=rotules)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(iris_data)
predictions = kmeans.predict(iris_data)

reduced_data = PCA(n_components=2).fit_transform(iris_data)

reduced_data_df = pd.DataFrame(reduced_data)
reduced_data_df[2] = predictions

# k-means clusters
fig, ax = plt.subplots()
ax.scatter(reduced_data_df[0], reduced_data_df[1], c=reduced_data_df[2])
plt.show()

# Original
fig, ax = plt.subplots()
ax.scatter(reduced_data_df[0], reduced_data_df[1], c=rotules)
plt.show()
```
