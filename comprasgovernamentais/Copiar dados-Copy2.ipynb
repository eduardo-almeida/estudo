```python
import pandas as pd
```

```python
def getLicitacao(inicio=0):
    #Se puxar todos dá erro, algumas linhas tem aspas duplas que prejudica a leitura
    data = pd.read_csv('D_CMPR_COMPRA.dat.gz',encoding='iso-8859-1', sep='¥', nrows= 200, skiprows= inicio)
    #data = pd.read_csv('D_CMPR_COMPRA.dat.gz',encoding='iso-8859-1', sep='¥', nrows= 42835, skiprows= inicio)
    data = data[['ID_UNDD_LOTACAO_PREGOEIRO', 'ID_CMPR_MODALIDADE_COMPRA', 'DS_CMPR_COMPRA']]
    data.rename(columns={'ID_UNDD_LOTACAO_PREGOEIRO':'Id do UASG', 
                         'ID_CMPR_MODALIDADE_COMPRA':'Id da Modalidade da Licitação', 
                         'DS_CMPR_COMPRA':'Identificador da Licitação'}, inplace = True)
    data = mergeUASG(data)
    data = mergeModalidadeLicitacao(data)
    return data
```

```python
def getListUASG():
    data = pd.read_csv('D_UNDD_UNIDADE.dat.gz',encoding='iso-8859-1', sep='¥')
    data = data[['ID_UNDD_UNIDADE','NO_UNDD_UNIDADE']]
    data.rename(columns={'ID_UNDD_UNIDADE':'Id do UASG','NO_UNDD_UNIDADE':'UASG'}, inplace = True)
    return data

def mergeUASG(licitacao): 
    lista = getListUASG()
    juncao = pd.merge(licitacao, lista, on='Id do UASG')
    juncao = juncao[['Id do UASG', 'UASG', 'Id da Modalidade da Licitação', 'Identificador da Licitação']]
    return juncao
```

```python
def getListModalidadeLicitacao():
    data = pd.read_csv('D_CMPR_MODALIDADE_COMPRA.dat.gz',encoding='iso-8859-1', sep='¥')
    data = data[['ID_CMPR_MODALIDADE_COMPRA','DS_CMPR_MODALIDADE_COMPRA']]
    data.rename(columns={'ID_CMPR_MODALIDADE_COMPRA':'Id da Modalidade da Licitação',
                         'DS_CMPR_MODALIDADE_COMPRA':'Modalidade da Licitação'}, 
                inplace = True)
    return data

def mergeModalidadeLicitacao(licitacao): 
    lista = getListModalidadeLicitacao()
    juncao = pd.merge(licitacao, lista, on='Id da Modalidade da Licitação')
    juncao = juncao[['Id do UASG', 'UASG', 'Id da Modalidade da Licitação', 'Modalidade da Licitação', 
                     'Identificador da Licitação']]
    juncao.rename(columns={'Id da Modalidade da Licitação':'Id ML'}, inplace = True)
    return juncao
```
