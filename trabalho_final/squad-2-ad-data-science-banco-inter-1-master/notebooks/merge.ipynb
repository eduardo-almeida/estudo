# Aplicando o Merge nas bases de dados

```python
#Importando pacotes
import pandas as pd
```

## Leitura dos arquivos

```python
df_portifolio_1 = pd.read_csv('../data/estaticos_portfolio1.csv')
```

```python
df_portifolio_2 = pd.read_csv('../data/estaticos_portfolio2.csv')
```

```python
df_portifolio_3 = pd.read_csv('../data/estaticos_portfolio3.csv')
```

```python
df_market = pd.read_csv('../data/estaticos_market.csv')
```

```python
df_merge_port_2 = df_market.merge(df_portifolio_2, on='id', how='inner')
```

```python
df_merge_port_3 = df_market.merge(df_portifolio_3, on='id', how='inner')
```

```python
df_merge_port_2.to_csv('../data/estaticos_portfolio2_merge.csv', index=False)
```

```python
df_merge_port_3.to_csv('../data/estaticos_portfolio3_merge.csv', index=False)
```
